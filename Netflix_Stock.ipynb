```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
import matplotlib.ticker as mtick
import seaborn as sns
```

```python
# Loading dataset
df = pd.read_csv(r"https://github.com/PrachiHolkar/Netflix_Stock/blob/main/NFLX_stocks.csv")
df
```

```python
# Removing header row accidentally read as data (row index 0)
df = df.drop(index = 0)
```

```python
# Converting 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])
```

```python
# Setting 'date' column as index
df.set_index('date', inplace = True)
```

```python
df
```

```python
df.info()
```

```python
# Ensuring correct data types
numeric_cols = ['open','high','low','close','adj_close','volume']
numeric_cols
```

```python
df[numeric_cols] = df[numeric_cols].astype(float)
```

```python
# Plotting closing price over time
plt.figure(figsize=(16,8))
plt.plot(df['close'], label='Closing Price', color='green')
plt.title('Netflix Closing Stock Price Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculating daily returns
df['daily_return'] = df['close'].pct_change()
df
```

```python
# Plotting daily returns
plt.figure(figsize=(12, 4))
plt.plot(df['daily_return'], color = 'green', label='Daily % Return')
plt.title('Netflix Daily Stock Returns Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Daily Return', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(1.0)) 
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Calculating moving averages
df['30_day_avg'] = df['close'].rolling(window=30).mean()
df['90_day_avg'] = df['close'].rolling(window=90).mean()
```

```python
# Plotting closing price with 30 and 90 day moving averages
plt.figure(figsize=(12, 8))
plt.plot(df['close'], label='Closing Price', alpha=0.5, color = 'green')
plt.plot(df['30_day_avg'], label='30-Day Average', color='red')
plt.plot(df['90_day_avg'], label='90-Day Average', color='blue')
plt.title('Netflix Stock Price with Moving Averages', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
yearly_return = df['daily_return'].resample('Y').mean()

# Plotting yearly returns 
colors = ['green' if val > 0 else 'red' for val in yearly_return]
plt.figure(figsize=(12, 5))
plt.bar(yearly_return.index.year, yearly_return * 100, color=colors) 
plt.title('Average Yearly Return', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Return (%)', fontsize=12)
plt.gca().yaxis.set_major_formatter(PercentFormatter(xmax=1))

plt.tight_layout()
plt.show()
```
